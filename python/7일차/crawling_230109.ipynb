## 웹크롤링 개요
- 인터넷에서 정보 수집 (데이터)
  + 유명 웹사이트에서 기사, 댓글, 이미지 등을 수집
  + DB에 저장
  + 법적인 측면 : 불법
  + 상업용으로 활용하면 무조건 처벌 / 연구목적 또는 교육목적이면 큰 문제가 안됨
- API를 활용해서 수집하는 방법
  + 이 방법을 적극 활용
  + 개발을 좀 더 쉽고, 편하게 할 수 있음
  + 단점 : 요금제한 / 사용 제한 / 출력 제한

##네이버 사이트 크롤링 간단 예제
- 상태코드 확인 : https://www.whatap.io/ko/blog/40/
- 200 : 정상적으로 작동
- 404 : 주소 입력이 잘못됨(주소가 사라지는 경우)
- 503 : 회사 서버가 내려간 상태


```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.naver.com/'

# url 변수에 담긴 url(네이버) html 문서를 출력
req = requests.get(url)
print(req.status_code)
print(req.text)
```

### 네이버 헤드 뉴스 찾기
- 1단계 : 사이트 정상인지 확인
- .py 만들고 배치로 돌려야 함.

```python
import requests
from bs4 import BeautifulSoup

# 데이터 수집
def crawling(soup):
  # print(soup)
  result = []

  # div 태그명 list_issue
  div = soup.find('div', class_ = 'list_issue')
  print(div)
  print('-----')

  for a in div.find_all('a'):
    print(a.get_text())
    result.append(a.get_text())

  return result

# 데이터 가공 (향후 프로젝트 진행 시)

# 데이터 DB 저장 (향후 프로젝트 진행 시)

def main():
  # 코드 작성 후, crawling 함수에 객체 전달 (soup)
  custom_header = {
      'referer' : 'https://www.naver.com/',
      'user-agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'
  }

  req = requests.get(url, headers = custom_header)

  # soup 객체
  soup = BeautifulSoup(req.text, "html.parser")
  result = crawling(soup)
  print('----최종 결과----')
  print(result)

if __name__ == "__main__":
  main()
```

## 벅스뮤직 음원차트 순위 추출

```python
import requests
from bs4 import BeautifulSoup

url = 'https://music.bugs.co.kr/chart'

req = requests.get(url)
print(req.status_code)
print(req.text)
```

```python
import requests
from bs4 import BeautifulSoup

def crawling(soup):
  #print(soup)
  result = []

  tbody = soup.find('tbody')
  print('-----')

  for p in tbody.find_all('p', class_ = 'title'):
    print(p.get_text())
    result.append(p.get_text().replace('\n', ''))

  return result

def main():
  custom_header = {
      'referer' : 'https://music.bugs.co.kr/chart',
      'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'
  }

  req = requests.get(url, headers = custom_header)

  soup = BeautifulSoup(req.text, "html.parser")
  result = crawling(soup)
  print('----최종 결과----')
  print(result)

if __name__ == "__main__":
  main()
```
